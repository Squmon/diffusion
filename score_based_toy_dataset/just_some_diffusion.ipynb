## Прямые процессы

```python
def forward_VPE(x0, index, betas):
    alphas = 1 - betas
    alphas_bar = tf.math.cumprod(alphas)
    return tf.sqrt(alphas_bar[index])*x0 + tf.sqrt(1 - alphas_bar[index])*tf.random.normal(tf.shape(x0))
```

## Сэплирование

```python
def langevin_sampling(x_shape, score_function, steps, noise_scale = 1.0):
    T = tf.linspace(1, 0, steps)
    dt = 1/steps
    x = tf.random.normal(x_shape)
    for t in T:
        x = x + dt*score_function(x = x, t = t) + noise_scale*tf.sqrt(2*dt) * tf.random.normal(x_shape)
    return x
```
